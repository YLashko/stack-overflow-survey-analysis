```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("survey_results_prepared.csv")
```

```python
languages = data.columns[list(data.columns).index("APL"):]
```

```python
data = data[languages]
```

```python
true_counts = data.apply(lambda r: r.value_counts()[True])
```

```python
columns_to_keep = true_counts[true_counts > 1500].index
columns_to_keep
```

```python
data = data[columns_to_keep]
```

```python
corr = data.corr()
corr.style.background_gradient(cmap='coolwarm')
```
