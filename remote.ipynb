```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
```

```python
data = pd.read_csv("survey_results_prepared.csv")
```

```python
data["RemoteWork"].unique()
```

```python
grouped = data.groupby(["RemoteWork"]).agg(["median"])["CompTotal"]
grouped.rename(index={"Hybrid (some remote, some in-person)": "Hybrid"}, inplace=True)
grouped.plot.bar(fontsize=11, title="Remote work salary compared to hybrid/in office")
plt.savefig(f"graphs/salary_remote/median.png")
```
